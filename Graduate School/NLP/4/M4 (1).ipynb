## POS Tagging

```python
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    print(pos_tags)
```

## Dependency Parsing

```python
for sentence in corpus:
    doc = nlp(sentence)
    displacy.render(doc)

```

## NER

```python
for sentence in corpus:
    doc = nlp(sentence)
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            structured_data["Name"] = ent.text
        elif ent.label_ == "ORG":
            structured_data["Organization"] = ent.text
        elif ent.label_ == "EDUCATION":
            structured_data["Education"].append(ent.text)
    print(structured_data)
```
