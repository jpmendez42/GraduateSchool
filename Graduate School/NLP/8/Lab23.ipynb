```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-en-es"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
English_OG = "This document is the ultimate authority for Magic: The Gathering® game play. It consists of a series of numbered rules followed by a glossary. Many of the numbered rules are divided into subrules, and each separate rule and subrule of the game has its own number. Changes may have been made to this document since its publication. You can download the most recent version from the Magic rules website at Magic.Wizards.com/Rules."

print("Original English Text:")
print(English_OG)
```

```python
inputs = tokenizer(English_OG, return_tensors="pt", padding=True, truncation=True)
```

```python
translated_tokens = model.generate(**inputs, max_length=100, num_beams=4)

translation = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

print("\nTransformer (Opus-MT) Translation:")
print(translation)
```
